````python
import torch
import matplotlib.pyplot as plt
from timeit import default_timer as timer
````

````python
# create a x tensor and A, B number
x = torch.arange(0,1,0.1)

b = 2
a = 2
````

````python
# Functions
def linear_function(a, b, x : torch.Tensor):
  return b + a*x



def plt_graph(x_position ,
              y_position ,
              predictions = None):
  plt.figure(figsize=(10,7))

  plt.scatter(x_position, y_position, c = "b", s = 4, label = "Graph")


  plt.legend(prop={"size": 10})
````

````python
# Calling the functions

y = linear_function(a, b, x)
print(y, y.shape)

# plotting the graph from our method
plt_graph(x, y)
````

````python
# non-linear function
def non_linear_function(b, x: torch.Tensor):
  return b + x + x**2
````

````python
y = non_linear_function(b, x)
print(y, y.shape)

plt_graph(x, y)
````

````python
# Measure a time difference between cuda and cpu

# Cpu tensors and operations
timer_start = timer()
g = torch.rand(100,100,200,200)
h = torch.rand(100,100,200,200)

mul_timer_start = timer()
k = torch.mul(g, h)
torch.mul(g, h)
torch.mul(g, h)
torch.mul(g, h)

timer_end = timer()

cpu_time = timer_end - timer_start
cpu_mul_time = timer_end - mul_timer_start

# GPU tensors and operations
timer_start = timer()
g = torch.rand(100,100,200,200).to("cuda")
h = torch.rand(100,100,200,200).to("cuda")

# measure just a multiplications
mul_timer_start = timer()
k = torch.mul(g, h)
torch.mul(g, h)
torch.mul(g, h)
torch.mul(g, h)

timer_end = timer()

gpu_time = timer_end - timer_start
gpu_mul_time = timer_end - mul_timer_start

# printout
print("CPU time: ", cpu_time)
print("CPU multiplication time: ", cpu_mul_time)
print("\nGPU time: ", gpu_time)
print("GPU multiplication time: ", gpu_mul_time)
print("\nTime difference: ", cpu_time - gpu_time)
````
